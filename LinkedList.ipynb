```python
class SinglyNode:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next        
```

```python
Head = SinglyNode(1)
A = SinglyNode(2)
B = SinglyNode(3)
C = SinglyNode(4)
D = SinglyNode(5)
```

```python
Head.next = A
A.next = B
B.next = C
C.next = D
```

```python
# Traverse the list in O(n)
curr = Head

while curr:
    print(curr.value)
    curr = curr.next
```

```python
# Display linked list in O(n)
def display(Head):
    curr = Head
    res = []
    while curr:
        res.append(str(curr.value))
        curr = curr.next
    print(("->".join(res)))
    
display(Head)
```

```python
def search(Head, num):
    curr = Head
    while curr:
        if num == curr.value:
            return True
        curr = curr.next
    return False
search(Head, 43)
```
