```python
class DoublyNode:
    def __init__(self, value, next=None, prev=None):
        self.value = value
        self.next = next
        self.prev = prev
        
    def __str__(self):
        return str(self.value)
```

```python
Head = Tail = DoublyNode(1)
print(Tail)
```

```python
# Display takes O(n)
def display(Head):
    curr = Head
    res = []
    while curr:
        res.append(str(curr.value))
        curr = curr.next
    print("<->".join(res))
    
display(Head)
```

```python
# Insert at beginning O(1)
def insert_at_beggining(Head, Tail, value):
    new_node = DoublyNode(value, next=Head)
    Head.prev = new_node
    return new_node, Tail

Head, Tail = insert_at_beggining(Head, Tail, 2)
display(Head)
```

```python
# Insert at end O(1)
def insert_at_end(Head, Tail, value):
    new_node = DoublyNode(value, next=None, prev=Tail)
    Tail.next = new_node
    return Head, new_node
Head, Tail = insert_at_end(Head, Tail, 4)
display(Head)
```
