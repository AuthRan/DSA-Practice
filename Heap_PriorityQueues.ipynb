```python
# Build Min-Heap (Heapify)
# Time : O(n), Space : O(1)
A = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]   # Random tree that doesn't satisfy heap

import heapq
heapq.heapify(A)
A 
# THe proces is simple select first non leaf node and shift down similarly move left twoards root and keep performing "Shift down" ie if parent node is smaller than leaf node then swap them
```

```python
# Heap push or pop
# Time : O(logn)
heapq.heappush(A, 4)
min = heapq.heappop(A)
A,min
```

```python
# Notic how this heap isn't sorted in array but yes its sorted if u make a tree of it
# As we are intreseted in sorted array we can use heapsort which is amongst the best sorting algorithm.
```

```python
# Heap Sort
# TIme : O(nlogn) Space : O(n)
# Note : O(1) space is possible via swapping, but this is complex

def heapsort(arr):
    heapq.heapify(arr)
    n = len(arr)
    new_list = [0] * n
    for i in range(n):
        min_value = heapq.heappop(arr)
        new_list[i] = min_value
    return new_list

heapsort(A)
```
